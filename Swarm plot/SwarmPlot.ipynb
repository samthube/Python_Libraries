```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset

```python
def totalbill_vs_size(data):
    """
    Description:
        This function plots a swarm plot of 'total_bill' against 'size'.

    Parameter:
        data: DataFrame containing the tips dataset.

    Return:
        None
    """
    sns.swarmplot(data=data, x="size", y="total_bill",size=4)
    
    plt.title('Swarm Plot of Total Bill vs. Size')
    plt.xlabel('Size')
    plt.ylabel('Total Bill')
    
    plt.show()

def main():
    try:
       
        data = pd.read_csv("D:\\Python Libraries\\Data\\tips.csv")
        
        totalbill_vs_size(data)
    except FileNotFoundError:
        print("File not found. Please check the file path.")

if __name__ == "__main__":
    main()
```
