```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw bar plot of sex against survived for a dataset

```python
def sex_vs_survived(data):
    """
    Description:
        This function plots a bar plot of 'sex' against 'survived'.

    Parameter:
        data: DataFrame containing the Titanic dataset.

    Return:
        None
    """
    sns.barplot(data=data, x="sex", y="survived")
    
    plt.title("Bar Plot of Sex Against Survived")
    plt.xlabel("Sex")
    plt.ylabel("Survived")
   
    plt.show()

def main():
    try:
        
        data = pd.read_csv("D:\\Python Libraries\\Data\\titanic.csv")
        
        sex_vs_survived(data)
    except FileNotFoundError:
        print("File not found. Please check the file path.")

if __name__ == "__main__":
    main()
```
