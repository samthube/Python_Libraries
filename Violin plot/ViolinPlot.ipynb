```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a violin plot of sex against total_bill for a given dataset.

```python
def sex_vs_total_bill(data):
    """
    Description:
        This function plots a violin plot of 'sex' against 'total_bill'.
    
    Parameter:
        data: DataFrame containing the tips dataset.
    
    Return:
        None
    """

    sns.violinplot(data=data, x="sex", y="total_bill")
  
    plt.title('Violin Plot of Sex vs. Total Bill')
    plt.xlabel('Sex')
    plt.ylabel('Total Bill')

    plt.show()

def main():
    try:
        data = pd.read_csv("D:\\Python Libraries\\Data\\tips.csv")
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return

    sex_vs_total_bill(data)

if __name__ == "__main__":
    main()
```

2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset

```python
def species_vs_sepal_length(data):
    """
    Description:
        This function plots a violin plot of 'species' against 'sepal_length'.
    
    Parameter:
        data: DataFrame containing the iris dataset.
    
    Return:
        None
    """
    sns.violinplot(data=data, x="species", y="sepal_length")
    
    plt.title('Violin Plot of Species vs. Sepal Length')
    plt.xlabel('Species')
    plt.ylabel('Sepal Length')
    
    plt.show()

def main():
    try:
        data = pd.read_csv("D:\\Python Libraries\\Data\\iris.csv")
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return
    species_vs_sepal_length(data)

if __name__ == "__main__":
    main()
```
