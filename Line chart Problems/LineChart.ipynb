1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
def plot_simple_line(x,y):
    """
    Description:
        Draws a simple line plot with the provided x and y axis values, including labels on the x-axis, y-axis, and a title.

    Parameter:
        x: List of x-axis values.
        y: List of y-axis values.

    Return:
        None
    """
    plt.plot(x, y, marker='o')
    plt.xlabel('X Axis Label')
    plt.ylabel('Y Axis Label')
    plt.title('Simple Line Plot')
    plt.show()

def main():
    
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]

    plot_simple_line(x,y)

if __name__ == "__main__":
    main()
```

3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.

```python
def plot_from_file():
    """
    Description:
        Draws a line plot using x and y axis values read from a text file, with labels on the x-axis, y-axis, and a title.

    Parameter:
        None

    Return:
        None
    """
    x_values = []
    y_values = []

    with open('test.txt', 'r') as file:
        for line in file:
            x, y = map(int, line.split())
            x_values.append(x)
            y_values.append(y)

    plt.plot(x_values, y_values, marker='o')
    plt.xlabel('X Axis Label')
    plt.ylabel('Y Axis Label')
    plt.title('Line Plot from File Data')
    plt.show()

def main():
    plot_from_file()

if __name__ == "__main__":
    main()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python
def plot_multiple_lines(x, y1, y2, y3):
    """
    Description:
        Draws multiple lines on the same plot using the provided x-axis values and three sets of y-axis values.
        Each line has a different marker and label.

    Parameter:
        x: List of x-axis values.
        y1: List of y-axis values for Line 1.
        y2: List of y-axis values for Line 2.
        y3: List of y-axis values for Line 3.

    Return:
        None
    """
    plt.plot(x, y1, label='Line 1', marker='o')
    plt.plot(x, y2, label='Line 2', marker='s')
    plt.plot(x, y3, label='Line 3', marker='^')

    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.title('Multiple Lines on Same Plot')
    plt.legend()
    plt.show()

def main():
    x = [1, 2, 3, 4, 5]
    y1 = [2, 4, 6, 8, 10]  
    y2 = [1, 3, 5, 7, 9]   
    y3 = [2, 3, 4, 5, 6]   
    plot_multiple_lines(x, y1, y2, y3)

if __name__ == "__main__":
    main()
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
def plot_and_adjust_axis_limits(x, y):
    """
    Description:
        Plots a line graph and displays the current axis limits.
        Then, sets new axis limits and updates the plot.

    Parameter:
        x: List of x-axis values.
        y: List of y-axis values.

    Return:
        None
    """
    plt.plot(x, y, marker='o')

    current_limits = plt.axis()
    print(f"Current axis limits: {current_limits}")

    plt.axis([0, 10, 0, 15])  
    
    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.title('Plot with Adjusted Axis Limits')
    
    plt.show()

def main():
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]
    plot_and_adjust_axis_limits(x, y)

if __name__ == "__main__":
    main()
```

12. Write a Python program to create multiple types of charts

```python
def create_charts(x, y, categories, values, scatter_x, scatter_y, histogram_data):
    """
    Description:
        Creates and displays multiple types of charts, including a line chart, bar chart,
        scatter plot, and histogram using the provided data.

    Parameter:
        x: List or array of x-axis values for the line chart.
        y: List or array of y-axis values for the line chart.
        categories: List of categories for the bar chart.
        values: List of values for the bar chart.
        scatter_x: List or array of x-axis values for the scatter plot.
        scatter_y: List or array of y-axis values for the scatter plot.
        histogram_data: Array or list of data for the histogram.

    Return:
        None
    """
  
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    axs[0, 0].plot(x, y, 'b-o')
    axs[0, 0].set_title('Line Chart')
    axs[0, 0].set_xlabel('X Axis')
    axs[0, 0].set_ylabel('Y Axis')

    axs[0, 1].bar(categories, values, color='orange')
    axs[0, 1].set_title('Bar Chart')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    axs[1, 0].scatter(scatter_x, scatter_y, color='green', marker='x')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('X Axis')
    axs[1, 0].set_ylabel('Y Axis')

    axs[1, 1].hist(histogram_data, bins=30, color='purple', edgecolor='black')
    axs[1, 1].set_title('Histogram')
    axs[1, 1].set_xlabel('Value')
    axs[1, 1].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()

def main():
   
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([2, 4, 6, 8, 10])
    categories = ['A', 'B', 'C', 'D']
    values = [10, 20, 15, 25]
    scatter_x = np.array([1, 2, 3, 4, 5])
    scatter_y = np.array([5, 15, 10, 20, 25])
    histogram_data = np.random.randn(1000) 

    create_charts(x, y, categories, values, scatter_x, scatter_y, histogram_data)

if __name__ == "__main__":
    main()
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python
def plot_closing_values(data):
    """
    Description:
        Draws a line chart of the closing values of Alphabet Inc. between October 3, 2016, 
        and October 7, 2016, and customizes the grid lines with major and minor grids.

    Parameter:
        data: DataFrame containing date and closing values.

    Return:
        None
    """
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Value', marker='o', color='blue')

    plt.xlabel('Date')
    plt.ylabel('Closing Value (USD)')
    plt.title('Alphabet Inc. Closing Values (Oct 3, 2016 - Oct 7, 2016)')

    plt.grid(True, which='both', linestyle='-', linewidth=0.5, color='blue')

    plt.minorticks_on()
    plt.grid(which='major', linestyle='-', linewidth=1, color='blue')

    plt.grid(which='minor', linestyle=':', linewidth=0.5, color='lightblue')

    plt.tick_params(axis='both', which='both', length=0)

    plt.legend()
    plt.xticks(rotation=45) 
    plt.tight_layout() 
    plt.show()

def main():
    
    data = pd.read_csv('closing_values.csv')

    data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%y')

    plot_closing_values(data)

if __name__ == "__main__":
    main()
```

15. Write a Python program to create multiple plots.

```python
def create_multiple_plots(x, y, categories, values, scatter_x, scatter_y, histogram_data):
    """
    Description:
        Creates multiple types of plots in a single figure, including a line plot, bar chart,
        scatter plot, and histogram, each in a different subplot.

    Parameter:
        x: List or array of x-axis values for the line plot.
        y: List or array of y-axis values for the line plot.
        categories: List of categories for the bar chart.
        values: List of values for the bar chart.
        scatter_x: List or array of x-axis values for the scatter plot.
        scatter_y: List or array of y-axis values for the scatter plot.
        histogram_data: Array or list of data for the histogram.

    Return:
        None
    """
    fig, axs = plt.subplots(2, 2, figsize=(14, 10))

    axs[0, 0].plot(x, y, 'b-o')
    axs[0, 0].set_title('Line Plot')
    axs[0, 0].set_xlabel('X Axis')
    axs[0, 0].set_ylabel('Y Axis')

    axs[0, 1].bar(categories, values, color='orange')
    axs[0, 1].set_title('Bar Chart')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    axs[1, 0].scatter(scatter_x, scatter_y, color='green', marker='x')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('X Axis')
    axs[1, 0].set_ylabel('Y Axis')

    axs[1, 1].hist(histogram_data, bins=30, color='purple', edgecolor='black')
    axs[1, 1].set_title('Histogram')
    axs[1, 1].set_xlabel('Value')
    axs[1, 1].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()

def main():
 
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([2, 4, 6, 8, 10])
    categories = ['A', 'B', 'C', 'D']
    values = [10, 20, 15, 25]
    scatter_x = np.array([1, 2, 3, 4, 5])
    scatter_y = np.array([5, 15, 10, 20, 25])
    histogram_data = np.random.randn(1000) 

    create_multiple_plots(x, y, categories, values, scatter_x, scatter_y, histogram_data)

if __name__ == "__main__":
    main()
```
