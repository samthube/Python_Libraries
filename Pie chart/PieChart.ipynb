```python
import matplotlib.pyplot as plt
import pandas as pd
```

1. Write a Python programming to create a pie chart of the popularity of programming
Languages.

```python
def create_pie_chart(languages, popularity):
    """
    Description:
        Creates a pie chart showing the popularity of different programming languages.

    Parameter:
        languages (list): List of programming languages.
        popularity (list): List of popularity percentages corresponding to the languages.
        
    Return:
        None
    """
   
    plt.pie(popularity, labels=languages, autopct='%1.1f%%')

    plt.show()

def main():
  
    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    create_pie_chart(languages, popularity)

if __name__ == "__main__":
    main()
```

2. Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.

```python
def create_pie_chart_with_title(languages, popularity):
    """
    Description:
        Creates a pie chart with a title showing the popularity of different programming languages.

    Parameter:
        languages (list): List of programming languages.
        popularity (list): List of popularity percentages corresponding to the languages.
        
    Return:
        None
    """
    
    plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)

    plt.title('Popularity of Programming Languages')

    plt.show()

def main():
 
    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    create_pie_chart_with_title(languages, popularity)

if __name__ == "__main__":
    main()
```

3. Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python
def gold_medal(df):
    """
    Description:
        Creates a pie chart showing the distribution of gold medals among countries.

    Parameter:
        df (DataFrame): The DataFrame containing country and gold medal data.
        
    Return:
        None
    """

    plt.pie(df["gold_medal"], labels=df["country"], autopct="%0.1f%%")

    plt.title('Gold Medal Achievements in 2016 Summer Olympics')

    plt.show()

def main():
    """
    Description:
        Reads the medal data from a CSV file and generates a pie chart.

    Parameter:
        None

    Return:
        None
    """

    df = pd.read_csv("D:/Python Libraries/medal.csv")
    
    gold_medal(df)

if __name__ == "__main__":
    main()
```
