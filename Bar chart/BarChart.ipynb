1. Write a Python programming to display a bar chart of the popularity of programming
Languages.

```python
def languages_popularity(languages, popularity):
    """
    Description:
        This function plots a bar chart showing the popularity of various programming languages.
    
    Parameter:
        languages (list): A list of programming languages.
        popularity (list): A list of corresponding popularity values.
    
    Return:
        None
    """
    plt.bar(languages, popularity)

    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')
    plt.title('Popularity of Programming Languages')
  
    plt.show()

def main():
    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    languages_popularity(languages, popularity)

if __name__ == "__main__":
    main()
```

7. Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.

```python
def plot_popularity_chart(languages, popularity):
    """
    Description:
        Plots a bar chart to display the popularity of programming languages.
    
    Parameter:
        languages (list): A list of programming languages.
        popularity (list): A list of popularity scores corresponding to the languages.
    
    Return:
        None
    """
    
    positions = np.arange(len(languages))

    plt.bar(positions, popularity)

    plt.xticks(positions, languages)
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')
   
    plt.show()

def main():
    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
    plot_popularity_chart(languages, popularity)

if __name__ == "__main__":
    main()
```

9. Write a Python programming to display a bar chart of the popularity of programming
Languages. Increase bottom margin.

```python
def plot_custom_bar_chart(languages, popularity):
    """
    Description:
        Plots a bar chart with increased bottom margin to display the popularity of programming languages.
    
    Parameter:
        languages (list): A list of programming languages.
        popularity (list): A list of popularity scores corresponding to the languages.
        bar_width (float): The width of each bar in the bar chart.
    
    Return:
        None
    """
    positions = np.arange(len(languages))

    plt.bar(positions, popularity, width=0.5, align='center', alpha=0.7, color='green')
    
    plt.subplots_adjust(bottom=0.3)

    plt.xticks(positions, languages)
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')

    plt.show()

def main():
    
    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
    
    plot_custom_bar_chart(languages, popularity)

if __name__ == "__main__":
    main()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.

```python
def plot_group_gender_barchart(groups, means_men, means_women):
    """
    Description:
        Plots a bar chart to display scores by group and gender (men and women) with multiple X values.
    
    Parameter:
        groups (list): A list of groups.
        means_men (list): A list of mean scores for men.
        means_women (list): A list of mean scores for women.
    
    Return:
        None
    """
    bar_width = 0.35 
    positions = np.arange(len(groups))  

    plt.bar(positions, means_men, bar_width, label='Men', color='blue', alpha=0.7)
    plt.bar(positions + bar_width, means_women, bar_width, label='Women', color='green', alpha=0.7)
    
    plt.xlabel('Groups')
    plt.ylabel('Scores')
    plt.title('Scores by Group and Gender')
    
    plt.xticks(positions + bar_width / 2, groups)

    plt.legend()

    plt.show()

def main():
    
    groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
    means_men = [22, 30, 35, 35, 26]
    means_women = [25, 32, 30, 35, 29]
    
    plot_group_gender_barchart(groups, means_men, means_women)

if __name__ == "__main__":
    main()
```

11. Write a Python program to create bar plot from a DataFrame.

```python
def plot_dataframe_barchart(df):
    """
    Description:
        Plots a bar chart from the DataFrame data.
    
    Parameter:
        df (DataFrame): A pandas DataFrame containing the data.
    
    Return:
        None
    """
 
    df.plot(kind='bar')

    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.title('Bar Plot from DataFrame')

    plt.show()

def main():
    data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 4],
    'c': [8, 3, 7, 6, 4],
    'd': [5, 4, 4, 4, 3],
    'e': [7, 2, 7, 8, 3],
    'f': [6, 6, 8, 6, 2]
}

    df = pd.DataFrame(data)
    plot_dataframe_barchart(df)

if __name__ == "__main__":
    main()
```

12. Write a Python program to create bar plots with error bars on the same figure.

```python
def plot_bar_with_error_bars(categories, means, std_devs):
    """
    Description:
        Plots a bar chart with error bars representing the standard deviation of mean velocities.
    
    Parameter:
        categories (list): A list of category labels.
        means (list): A list of mean velocities.
        std_devs (list): A list of standard deviations for the mean velocities.
    
    Return:
        None
    """
    positions = np.arange(len(categories))
    bar_width = 0.6  
 
    plt.bar(positions, means, bar_width, yerr=std_devs, capsize=5, color='skyblue', alpha=0.7, ecolor='black', linestyle='--')

    plt.xlabel('Categories')
    plt.ylabel('Mean Velocity')
    plt.title('Mean Velocity with Error Bars')
    
    plt.xticks(positions, categories)
 
    plt.show()

def main():
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
    categories = ['A', 'B', 'C', 'D']
    plot_bar_with_error_bars(categories, mean_velocity, std_dev_velocity)

if __name__ == "__main__":
    main()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).

```python
def plot_bar_with_error_bars_and_labels(categories, means, std_devs):
    """
    Description:
        Plots a bar chart with error bars and adds text labels above each bar displaying the integer values of the mean velocities.
    
    Parameter:
        categories (list): A list of category labels.
        means (list): A list of mean velocities.
        std_devs (list): A list of standard deviations for the mean velocities.
    
    Return:
        None
    """
    positions = np.arange(len(categories))  
    bar_width = 0.6  
   
    bars = plt.bar(positions, means, bar_width, yerr=std_devs, capsize=5, color='skyblue', alpha=0.7, ecolor='black', linestyle='--')

    for bar, mean in zip(bars, means):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{int(mean * 1000)}', ha='center', va='bottom')

    plt.xlabel('Categories')
    plt.ylabel('Mean Velocity')
    plt.title('Mean Velocity with Error Bars')
   
    plt.xticks(positions, categories)
    
    plt.show()

def main():
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
    categories = ['A', 'B', 'C', 'D']
    plot_bar_with_error_bars_and_labels(categories, mean_velocity, std_dev_velocity)

if __name__ == "__main__":
    main()
```
