```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw box plot of life expectancy by continent for a data set

```python
def lifeExp_by_continent(data):
    """
    Description:
        This function plots a box plot of life expectancy by continent.

    Parameter:
        data: DataFrame containing the gapminder dataset.

    Return:
        None
    """
    sns.boxplot(data=data, x="continent", y="lifeExp")
    
    plt.title('Box Plot of Life Expectancy by Continent')
    plt.xlabel('Continent')
    plt.ylabel('Life Expectancy')
    
    plt.show()

def main():
    try:
        data = pd.read_csv("D:\\Python Libraries\\Data\\gapminder-FiveYearData.csv")
        
        lifeExp_by_continent(data)
    except FileNotFoundError:
        print("File not found. Please check the file path.")

if __name__ == "__main__":
    main()
```

2. Write a program to draw a box plot of day by tips for a dataset

```python
def day_vs_tip(data):
    """
    Description:
        This function plots a box plot of day against tip.

    Parameter:
        data: DataFrame containing the tips dataset.

    Return:
        None
    """
    sns.boxplot(data=data, x="day", y="tip")
    
    plt.title('Box Plot of Day vs. Tip')
    plt.xlabel('Day')
    plt.ylabel('Tip')
    
    plt.show()

def main():
    try:
        data = pd.read_csv("D:\\Python Libraries\\Data\\tips.csv")
       
        day_vs_tip(data)
    except FileNotFoundError:
        print("File not found. Please check the file path.")

if __name__ == "__main__":
    main()
```
