```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a point plot for sex against survived for a dataset

```python
def plot_survival_rate(data):
    """
    Description:
        This function plots a point plot for survival rate of passengers based on their sex.

    Parameter:
        data: DataFrame containing the Titanic dataset.

    Return:
        None
    """
 
    sns.pointplot(data=data, x="sex", y="survived")

    plt.title('Survival Rate by Sex on Titanic')

    plt.show()

def main():
   
    data = pd.read_csv("D:\\Python Libraries\\Data\\titanic.csv")

    plot_survival_rate(data)

if __name__ == "__main__":
    main()
```
