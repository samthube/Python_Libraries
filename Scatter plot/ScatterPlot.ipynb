```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset 

```python
def scatter_plot(data):
    """
    Description:
        This function plots a scatter plot of 'day' against 'total bill'.

    Parameter:
        data: DataFrame containing the tips dataset.

    Return:
        None
    """
    sns.scatterplot(data=data, x="day", y="total_bill")
  
    plt.title('Scatter Plot of Day vs. Total Bill')
    plt.xlabel('Day of the Week')
    plt.ylabel('Total Bill')
  
    plt.show()

def main():
    
    data = pd.read_csv("D:\\Python Libraries\\Data\\tips.csv")
    
    scatter_plot(data)

if __name__ == "__main__":
    main()
```
